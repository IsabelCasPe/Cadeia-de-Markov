```python
import numpy as np
import matplotlib.pyplot as plt

def newton_method_multidim(f, jacobian, hessian, x0, tol=1e-10, max_iter=10):
    """Aplica o método de Newton para encontrar um ponto crítico de uma função multivariável.

    Args:
        f: Função cuja derivada buscamos zerar.
        jacobian: Gradiente da função.
        hessian: Hessiana da função.
        x0: Chute inicial (vetor).
        tol: Tolerância para o critério de parada.
        max_iter: Número máximo de iterações.

    Returns:
        x: Aproximação do ponto crítico.
        history: Lista com os valores de x a cada iteração.
    """
    x = np.array(x0, dtype=float)
    history = [x.copy()]

    for _ in range(max_iter):
        grad = np.array(jacobian(x))
        hess = np.array(hessian(x))

        if np.linalg.norm(grad) < tol:
            break

        delta_x = np.linalg.solve(hess, -grad)
        x_new = x + delta_x
        history.append(x_new.copy())

        if np.linalg.norm(delta_x) < tol:
            break

        x = x_new

    return x, history

# Definição da função f(x, y) = x^2 + y^2 - 4x - 6y + 9 e suas derivadas
f = lambda x: x[0]**2 + x[1]**2 - 4*x[0] - 6*x[1] + 9
jacobian = lambda x: [2*x[0] - 4, 2*x[1] - 6]
hessian = lambda x: [[2, 0], [0, 2]]

# Chute inicial
x0 = [3.0, 3.0]

# Aplicar o método de Newton
critical_point, history = newton_method_multidim(f, jacobian, hessian, x0)

# Determinar se é máximo ou mínimo
hessian_matrix = np.array(hessian(critical_point))
eigenvalues = np.linalg.eigvals(hessian_matrix)

if np.all(eigenvalues > 0):
    nature = "mínimo local"
elif np.all(eigenvalues < 0):
    nature = "máximo local"
else:
    nature = "ponto de sela"

# Plotando a convergência (apenas coordenadas x e y)
history = np.array(history)
plt.figure(figsize=(8, 5))
plt.plot(history[:, 0], history[:, 1], marker='o', linestyle='-', color='b', label='Iterações')
plt.scatter(*critical_point, color='r', label=f'Ponto crítico: {critical_point} ({nature})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Convergência do Método de Newton para pontos críticos (2D)')
plt.legend()
plt.grid()
plt.show()

print(f"Ponto crítico encontrado: {critical_point} ({nature})")
```

O algoritmo de Metropolis-Hastings é um método de Amostragem de Cadeia de Markov de Monte Carlo (MCMC)

```python
import numpy as np
import matplotlib.pyplot as plt

def compute_spectral_gap(transition_matrix):
    """Calcula o buraco espectral de uma cadeia de Markov.

    Args:
        transition_matrix: Matriz de transição da cadeia de Markov.

    Returns:
        spectral_gap: Diferença entre o maior e o segundo maior autovalor em módulo.
    """
    eigenvalues = np.linalg.eigvals(transition_matrix)
    eigenvalues = np.sort(np.abs(eigenvalues))[::-1]  # Ordena em ordem decrescente
    spectral_gap = eigenvalues[0] - eigenvalues[1]
    return spectral_gap, eigenvalues

def gradient_descent(P, learning_rate=0.01, max_iter=100):
    """Aplica o algoritmo de descida de gradiente para otimizar a matriz de transição.

    Args:
        P: Matriz de transição inicial da cadeia de Markov.
        learning_rate: Taxa de aprendizado.
        max_iter: Número máximo de iterações.

    Returns:
        optimized_P: Matriz de transição otimizada.
        history: Lista de valores do buraco espectral durante a otimização.
    """
    P = np.array(P, dtype=float)
    history = []

    for _ in range(max_iter):
        spectral_gap, _ = compute_spectral_gap(P)
        history.append(spectral_gap)

        # Gradiente simples baseado na maximização do buraco espectral
        grad = np.random.randn(*P.shape) * 0.01  # Perturbação pequena
        P += learning_rate * grad
        P = np.abs(P)  # Garantir valores positivos
        P /= P.sum(axis=1, keepdims=True)  # Normalizar para manter a propriedade de matriz de transição

    return P, history

def generate_random_markov_matrix(size):
    """Gera uma matriz de transição aleatória para uma cadeia de Markov."""
    P = np.random.rand(size, size)
    return P / P.sum(axis=1, keepdims=True)

def metropolis_hastings(target_distribution, proposal_std=1.0, num_samples=10000):
    """Executa o algoritmo de Metropolis-Hastings para amostragem de uma distribuição alvo.

    Args:
        target_distribution: Função que define a distribuição alvo (não normalizada).
        proposal_std: Desvio padrão da proposta gaussiana.
        num_samples: Número de amostras desejadas.

    Returns:
        samples: Amostras geradas.
    """
    samples = []
    x = np.random.randn()  # Estado inicial

    for _ in range(num_samples):
        x_proposed = x + np.random.normal(0, proposal_std)  # Proposta gaussiana
        acceptance_ratio = min(1, target_distribution(x_proposed) / target_distribution(x))

        if np.random.rand() < acceptance_ratio:
            x = x_proposed

        samples.append(x)

    return np.array(samples)

# Testar diferentes matrizes de transição
P1 = generate_random_markov_matrix(3)
P2 = generate_random_markov_matrix(4)

# Aplicar o algoritmo em diferentes matrizes
optimized_P1, history1 = gradient_descent(P1)
optimized_P2, history2 = gradient_descent(P2)

# Comparação do buraco espectral em diferentes matrizes
spectral_gap1, _ = compute_spectral_gap(optimized_P1)
spectral_gap2, _ = compute_spectral_gap(optimized_P2)

# Aplicar o algoritmo de Metropolis-Hastings para uma distribuição Gaussiana
normal_target = lambda x: np.exp(-x**2 / 2)  # Distribuição Gaussiana não normalizada
samples = metropolis_hastings(normal_target)

# Plotando histograma das amostras
plt.figure(figsize=(8, 5))
plt.hist(samples, bins=50, density=True, alpha=0.6, color='g', label='Amostras de Metropolis-Hastings')
x_vals = np.linspace(-4, 4, 1000)
plt.plot(x_vals, normal_target(x_vals) / np.max(normal_target(x_vals)), 'r-', label='Distribuição Alvo')
plt.xlabel('Valor Amostrado')
plt.ylabel('Densidade')
plt.title('Amostragem com Algoritmo de Metropolis-Hastings')
plt.legend()
plt.grid()
plt.show()

# Plotando a evolução da convergência para diferentes matrizes
plt.figure(figsize=(8, 5))
plt.plot(history1, marker='o', linestyle='-', label='Matriz 3x3')
plt.plot(history2, marker='s', linestyle='-', label='Matriz 4x4')
plt.xlabel('Iteração')
plt.ylabel('Buraco Espectral')
plt.title('Comparação da Convergência em Diferentes Matrizes de Markov')
plt.legend()
plt.grid()
plt.show()

print(f"Buraco espectral otimizado para matriz 3x3: {spectral_gap1}")
print(f"Buraco espectral otimizado para matriz 4x4: {spectral_gap2}")
```
